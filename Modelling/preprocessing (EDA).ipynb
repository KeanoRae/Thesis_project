## Univariate Analysis

```python
df['Year Started in the Program'].value_counts()
```

```python
fig = plt.figure(figsize=(15,5))
sns.countplot(data=df, x=df['Year Started in the Program'], hue=df['Enrolled in 2nd Year'])
```

```python
fig = plt.figure(figsize=(15,5))
sns.countplot(data=df, x=df['Year Started in the Program'], hue=df['Gender'])
```

```python
fig = plt.figure(figsize=(15,5))
sns.countplot(data=df, x=df['Gender'], hue=df['Enrolled in 2nd Year'])
```

```python
fig = plt.figure(figsize=(15,5))
df['What track/strand did you take in SHS?'].value_counts().plot(kind='pie', title='SHS Strand', autopct='%1.0f%%')
```

```python
fig = plt.figure(figsize=(15,5))
sns.countplot(data=df, x=df['What track/strand did you take in SHS?'], hue=df['Enrolled in 2nd Year'])
```

```python
fig = plt.figure(figsize=(15,5))
sns.countplot(data=df, x=df['Did you have a computer or laptop before admission to CS/IT?'], hue=df['Enrolled in 2nd Year'])
```

```python
sns.countplot(data=df, x=df['CC100'], hue=df['Year Started in the Program'])
```

```python
CC100_year = df.groupby(['CC100', 'Year Started in the Program']).size().reset_index().pivot(columns='Year Started in the Program', index='CC100', values=0)
CC101_year = df.groupby(['CC101', 'Year Started in the Program']).size().reset_index().pivot(columns='Year Started in the Program', index='CC101', values=0)
CC102_year = df.groupby(['CC102', 'Year Started in the Program']).size().reset_index().pivot(columns='Year Started in the Program', index='CC102', values=0)
CSIT_year = df.groupby(['CS111/IT112', 'Year Started in the Program']).size().reset_index().pivot(columns='Year Started in the Program', index='CS111/IT112', values=0)
```

```python

sns.lineplot(data=CC100_year, dashes=False)
```

```python
sns.lineplot(data=CC101_year, dashes=False)
```

```python
sns.lineplot(data=CC102_year, dashes=False)
```

```python
sns.lineplot(data=CSIT_year, dashes=False)
```
